```python
import openai
import os
openai.api_key = os.getenv("OPENAI_API_KEY")

from openai import OpenAI
from pathlib import Path


```

# 创建speech

```python
speech_file_path = Path('__file__').parent / "speech.mp3"
speech_file_path
```

```python
response = openai.audio.speech.create(
    model = "tts-1", # One of the available TTS models: tts-1 or tts-1-hd
    voice="echo", # alloy, echo, fable, onyx, nova, and shimmer.
    input="The quick brown fox jumped over the lazy dog." # The text to generate audio for. The maximum length is 4096 characters.
)

response.stream_to_file(speech_file_path)
```

# 创建 transcription

```python
from openai import OpenAI
client = OpenAI()

audio_file = open("speech.mp3", "rb")

transcript = client.audio.transcriptions.create(
    model="whisper-1", # ID of the model to use. Only whisper-1 is currently available.
    file=audio_file
)
```

# 创建 translation

```python
from openai import OpenAI
client = OpenAI()

audio_file = open("speech.mp3", "rb")
transcript = client.audio.translations.create(
  model="whisper-1",  # ID of the model to use. Only whisper-1 is currently available.
  file=audio_file # The audio file object (not file name) translate, in one of these formats: flac, mp3, mp4, mpeg, mpga, m4a, ogg, wav, or webm.
)
```
